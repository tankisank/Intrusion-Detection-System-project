# GLCM 
## (for 1 image)

```python
image_spot = cv2.imread('AB1.jpg')
gray = cv2.cvtColor(image_spot, cv2.COLOR_BGR2GRAY)

# Parameters:
#   source image
#   List of pixel pair distance offsets - here 1 in each direction
#   List of pixel pair angles in radians
graycom = feature.greycomatrix(gray, [1], [0, np.pi/4, np.pi/2, 3*np.pi/4], levels=256)

# Find the GLCM properties
contrast = feature.greycoprops(graycom, 'contrast')
dissimilarity = feature.greycoprops(graycom, 'dissimilarity')
homogeneity = feature.greycoprops(graycom, 'homogeneity')
energy = feature.greycoprops(graycom, 'energy')
correlation = feature.greycoprops(graycom, 'correlation')
ASM = feature.greycoprops(graycom, 'ASM')

print("Contrast: {}".format(contrast))
print("Dissimilarity: {}".format(dissimilarity))
print("Homogeneity: {}".format(homogeneity))
print("Energy: {}".format(energy))
print("Correlation: {}".format(correlation))
print("ASM: {}".format(ASM))
```

## GLCM for multiple images

```python
images = [cv2.imread('AB1.jpg'),cv2.imread('AB2.jpg'),cv2.imread('AB3.jpg'),
          cv2.imread('AB4.jpg'),cv2.imread('AB5.jpg')]

glcmdf = pd.DataFrame()

for i in range(len(images)):
    gray = cv2.cvtColor(images[i], cv2.COLOR_BGR2GRAY)
    graycom = feature.greycomatrix(gray, [1], [0, np.pi/4, np.pi/2, 3*np.pi/4], levels=256)

    # Find the GLCM properties
    contrast = feature.greycoprops(graycom, 'contrast')
    dissimilarity = feature.greycoprops(graycom, 'dissimilarity')
    homogeneity = feature.greycoprops(graycom, 'homogeneity')
    energy = feature.greycoprops(graycom, 'energy')
    correlation = feature.greycoprops(graycom, 'correlation')
    ASM = feature.greycoprops(graycom, 'ASM')

    print("\n\nImage no. {}".format(i+1))
    print("Contrast: {}".format(contrast))
    print("Dissimilarity: {}".format(dissimilarity))
    print("Homogeneity: {}".format(homogeneity))
    print("Energy: {}".format(energy))
    print("Correlation: {}".format(correlation))
    print("ASM: {}".format(ASM))

    a0,a45,a90,a135 = contrast[0]
    Contrast = {'contrast 0° at 1 ':a0,'contrast 45° at 1 ':a45,'contrast 90° at 1 ':a90,'contrast 135° at 1 ':a135} 
    a0,a45,a90,a135 = dissimilarity[0]
    Dissimilarity = {'dissimilarity 0° at 1 ':a0,'dissimilarity 45° at 1 ':a45,'dissimilarity 90° at 1 ':a90,'dissimilarity 135° at 1 ':a135} 
    a0,a45,a90,a135 = homogeneity[0]
    Homogeneity = {'homogeneity 0° at 1 ':a0,'homogeneity 45° at 1 ':a45,'homogeneity 90° at 1 ':a90,'homogeneity 135° at 1 ':a135} 
    a0,a45,a90,a135 = energy[0]
    Energy = {'energy 0° at 1 ':a0,'energy 45° at 1 ':a45,'energy 90° at 1 ':a90,'energy 135° at 1 ':a135} 
    a0,a45,a90,a135 = correlation[0]
    Correlation = {'correlation 0° at 1 ':a0,'correlation 45° at 1 ':a45,'correlation 90° at 1 ':a90,'correlation 135° at 1 ':a135} 
    a0,a45,a90,a135 = ASM[0]
    asm = {'ASM 0° at 1':a0,'ASM 45° at 1':a45,'ASM 90° at 1':a90,'ASM 135° at 1':a135}

    glcm = {**Contrast,**Dissimilarity, **Homogeneity, **Energy, **Correlation,**asm}
    value = pd.DataFrame(glcm,index=[i])
    glcmdf = pd.concat([glcmdf,value],axis=0)
```

# GLRL 

## (for 1 image)

```python
from glrlm import GLRLM
import cv2

IMG = "AB1.jpg"
img = cv2.imread(IMG)

app = GLRLM()
glrlm = app.get_features(img, 8)
```

## Long Runs Emphasis

LRE=∑i∈Ng∑j∈Nrj2Pij/∑i∈Ng∑j∈NrPij

## Short Runs Emphasis

SRE=∑i∈Ng∑j∈NrPijj2/∑i∈Ng∑j∈NrPij

## Gray Level Non- Uniformaity

GLN=∑i∈Ng(∑j∈NrPij)2/∑i∈Ng∑j∈NrPij

## Run Length Non-uniformity

RLN=∑j∈Nr(∑i∈NgPij)2/∑i∈Ng∑j∈NrPij

## Run Percentage

RP=∑i∈Ng∑j∈NrPij/N

```python
print(glrlm.SRE,glrlm.LRE,glrlm.GLU,glrlm.RLU,glrlm.RPC)
```

## (for all images)

```python
# SRE = Short Run Emphasis
# LRE = Long Run Emphasis
# GLU = Grey Level Uniformity
# RLU = Run Length Uniformity
# RPC = Run Percentage
```

```python

images = [cv2.imread('AB1.jpg'),cv2.imread('AB2.jpg'),cv2.imread('AB3.jpg'),
          cv2.imread('AB4.jpg'),cv2.imread('AB5.jpg')]

glrlmdf = pd.DataFrame()

for i in range(len(images)):

    img = images[i]

    app = GLRLM()
    glrlm = app.get_features(img, 8)

    # Find the GLRLM properties
    sre = glrlm.SRE
    lre = glrlm.LRE
    glu = glrlm.GLU
    rlu = glrlm.RLU
    rpc = glrlm.RPC 

# SRE = Short Run Emphasis
# LRE = Long Run Emphasis
# GLU = Grey Level Uniformity
# RLU = Run Length Uniformity
# RPC = Run Percentage
    print("\n\nImage no. {}".format(i+1))
    print("Short Run Emphasis: {}".format(sre))
    print("Long Run Emphasis: {}".format(lre))
    print("Grey Level Uniformity: {}".format(glu))
    print("Run Length Uniformity: {}".format(rlu))
    print("Run Percentage: {}".format(rpc))

    
    features = {'SRE':sre,'LRE':lre,'GLU':glu,'RLU':rlu,'RPC':rpc,}
    value = pd.DataFrame(features,index=[i])
    glrlmdf = pd.concat([glrlmdf,value],axis=0)
```
